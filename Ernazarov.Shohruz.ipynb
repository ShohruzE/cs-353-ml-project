```python
# Create helper function to classify blood pressure for every row
def classify_bp(row):
    systolic_bp = row["BPXOSY1"]
    diastolic_bp = row["BPXODI1"]
    # Edge case where either systolic or diastolic are missing
    if pd.isna(systolic_bp) or pd.isna(diastolic_bp):
        return np.nan
    if systolic_bp < 120 and diastolic_bp < 80:
        return "normal"
    else:
        return "abnormal"
```

```python
# Pass in function to every row to classify and label the new BP column
nhanes['BP'] = nhanes.apply(classify_bp, axis=1)
nhanes['BP'].value_counts()
```

```python
nhanes['BP']
```

```python
def classify_cholesterol(value):
    if pd.isna(value):
        return np.nan
    if value < 200:
        return "normal"
    elif value <= 239:
        return "borderline"
    else:
        return "high"
```

```python
nhanes["CHOL"] = nhanes["LBXTC"].apply(classify_cholesterol)
nhanes["CHOL"].value_counts()
```

```python
nhanes = nhanes.drop(columns=['BPXOSY1', 'BPXODI1', 'LBXTC'])
```

### Task 5: 

```python
def pre_clean_data(df):
    
    # HIQ032 checkbox columns: NaN means "not checked" → 0, any value → 1
    hiq_cols = ['HIQ032A', 'HIQ032B', 'HIQ032D', 'HIQ032F', 'HIQ032H', 'HIQ032I']
    for col in hiq_cols:
        df[col] = df[col].notna().astype(int)
        
    # Remap binary 1/2 features to 1/0 (1=Yes stays 1, 2=No becomes 0)
    binary_remap = ['RIAGENDR', 'DMQMILIZ', 'DMDBORN4', 'ALQ111', 'ALQ151','HIQ011', 'HIQ210', 'INQ300']
    for col in binary_remap:
        df[col] = df[col].map({1: 1, 2: 0})

    # Drop non-predictive columns
    df = df.drop(columns=['SEQN', 'SDDSRVYR', 'RIDSTATR'])
    
    return df
```

```python
bp_clean = pre_clean_data(bp_training_ernazarov)
chol_clean = pre_clean_data(chol_training_ernazarov)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer

nominal_cols = ['RIDRETH1', 'DMDMARTZ']
numeric_cols = [col for col in bp_clean.drop(columns=['BP', 'CHOL']).columns if col not in nominal_cols]

numeric_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

nominal_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

preprocessor_b = ColumnTransformer(transformers=[
    ('numeric', numeric_pipeline, numeric_cols),
    ('nominal', nominal_pipeline, nominal_cols)
])

preprocessor_c = ColumnTransformer(transformers=[
    ('numeric', numeric_pipeline, numeric_cols),
    ('nominal', nominal_pipeline, nominal_cols)
])

plineb = Pipeline(steps=[('preprocessor', preprocessor_b)])
plinec = Pipeline(steps=[('preprocessor', preprocessor_c)])
```

```python
bp_X = bp_clean.drop(columns=['BP', 'CHOL'])
bp_y = bp_clean['BP']

bp_mask = bp_y.notna()
bp_X = bp_X[bp_mask]
bp_y = bp_y[bp_mask]

bp_transformed = plineb.fit_transform(bp_X)

bp_feature_names = plineb.named_steps['preprocessor'].get_feature_names_out()
bp_transformed_df = pd.DataFrame(bp_transformed, columns=bp_feature_names)
bp_transformed_df.tail()
```

```python
chol_X = chol_clean.drop(columns=['BP', 'CHOL'])
chol_y = chol_clean['CHOL']

chol_mask = chol_y.notna()
chol_X = chol_X[chol_mask]
chol_y = chol_y[chol_mask]

chol_transformed = plinec.fit_transform(chol_X)

chol_feature_names = plinec.named_steps['preprocessor'].get_feature_names_out()
chol_transformed_df = pd.DataFrame(chol_transformed, columns=chol_feature_names)
chol_transformed_df.tail()
```

### Note:
After running the algorithms with little-to-no hyperparameter tuning or default configurations, I decided to try a hyperparameter tuning method such as `GridSearch` to find the ideal hyperparamters for all of the algorithms to see if we could bump up our scores by a little bit.

```python
from sklearn.model_selection import GridSearchCV

param_grids = {
    'Decision Stump': {
        'classifier__criterion': ['gini', 'entropy'],
        'classifier__splitter': ['best', 'random']
    },
    'Decision Tree': {
        'classifier__max_depth': [3, 5, 10, 15, 20, None],
        'classifier__criterion': ['gini', 'entropy'],
        'classifier__min_samples_split': [2, 5, 10],
        'classifier__min_samples_leaf': [1, 2, 5]
    },
    'Random Forest': {
        'classifier__n_estimators': [100, 500, 1000],
        'classifier__max_depth': [5, 10, 20, None],
        'classifier__min_samples_split': [2, 5, 10],
        'classifier__min_samples_leaf': [1, 2, 5]
    },
    'Gradient Boosting': {
        'classifier__n_estimators': [100, 500, 1000],
        'classifier__learning_rate': [0.01, 0.1, 0.2],
        'classifier__max_depth': [3, 5, 7],
        'classifier__min_samples_split': [2, 5, 10]
    },
    'Support Vector Machine': {
        'classifier__C': [0.1, 1, 10, 100],
        'classifier__kernel': ['rbf', 'poly'],
        'classifier__gamma': ['scale', 'auto']
    },
    'Naive Bayes': {
        'classifier__var_smoothing': [1e-9, 1e-8, 1e-7, 1e-6, 1e-5]
    },
    'Logistic Regression': {
        'classifier__C': [0.01, 0.1, 1, 10, 100],
        'classifier__solver': ['lbfgs', 'saga'],
        'classifier__max_iter': [1000, 2000]
    },
    'Stochastic Gradient Descent': {
        'classifier__loss': ['log_loss', 'modified_huber'],
        'classifier__alpha': [0.0001, 0.001, 0.01],
        'classifier__penalty': ['l2', 'l1', 'elasticnet'],
        'classifier__max_iter': [1000, 2000]
    }
}
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.model_selection import cross_validate
from scipy.stats import f_oneway

algorithms = {
    "Decision Stump": DecisionTreeClassifier(max_depth=1, random_state=42),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(n_estimators=100, random_state=42),
    "Support Vector Machine": SVC(kernel='rbf', random_state=42),
    "Naive Bayes": GaussianNB(),
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "Stochastic Gradient Descent": SGDClassifier(random_state=42)
}

scoring = ['accuracy', 'f1_weighted', 'recall_weighted']
```

```python
bp_best_hyperparameters = {}

for name, clf in algorithms.items():
    pipe = Pipeline([
        ('preprocessor', preprocessor_b),
        ('classifier', clf)
    ])
    
    grid_search = GridSearchCV(
        pipe,
        param_grids[name],
        cv=5,
        scoring='f1_weighted',
        n_jobs=-1
    )
    grid_search.fit(bp_X, bp_y)
    
    bp_best_hyperparameters[name] = {
        'best_params': grid_search.best_params_,
        'best_score': round(grid_search.best_score_, 4)
    }
    print(f"{name}: {grid_search.best_score_:.4f} - {grid_search.best_params_}")

pd.DataFrame({name: info for name, info in bp_best_hyperparameters.items()}).T
```

```python
bp_algorithms = {
    'Decision Stump': DecisionTreeClassifier(max_depth=1, criterion='gini', splitter='best', random_state=42),
    'Decision Tree': DecisionTreeClassifier(criterion='gini', max_depth=3, min_samples_leaf=1, min_samples_split=2, random_state=42),
    'Random Forest': RandomForestClassifier(max_depth=10, min_samples_leaf=2, min_samples_split=10, n_estimators=500, random_state=42),
    'Gradient Boosting': GradientBoostingClassifier(learning_rate=0.01, max_depth=3, min_samples_split=2, n_estimators=500, random_state=42),
    'Support Vector Machine': SVC(C=1, gamma='scale', kernel='rbf', random_state=42),
    'Naive Bayes': GaussianNB(var_smoothing=1e-05),
    'Logistic Regression': LogisticRegression(C=1, max_iter=1000, solver='lbfgs', random_state=42),
    'Stochastic Gradient Descent': SGDClassifier(alpha=0.01, loss='log_loss', max_iter=1000, penalty='l1', random_state=42)
}
```

```python
bp_cv_results = {}

for name, model in bp_algorithms.items():
    pipeline = Pipeline(
        steps=[
            ('preprocessor', preprocessor_b), 
            ('classifier', model)
        ])
    cv_results = cross_validate(pipeline, bp_X, bp_y, cv=10, scoring=scoring)
    bp_cv_results[name] = cv_results
    
bp_cv_results
```

```python
bp_summary = pd.DataFrame({
    name: {
        'Accuracy (mean)': cv['test_accuracy'].mean().round(4),
        'Accuracy (std)': cv['test_accuracy'].std().round(4),
        'F1 Weighted (mean)': cv['test_f1_weighted'].mean().round(4),
        'F1 Weighted (std)': cv['test_f1_weighted'].std().round(4),
        'Recall Weighted (mean)': cv['test_recall_weighted'].mean().round(4),
        'Recall Weighted (std)': cv['test_recall_weighted'].std().round(4),
    }
    for name, cv in bp_cv_results.items()
}).T # swap rows and columns

bp_summary
```

```python
bp_accuracy_by_algorithm = [bp_cv_results[name]['test_accuracy'] for name in algorithms]

f_stat, p_value = f_oneway(*bp_accuracy_by_algorithm)
print(f"BP ANOVA: F-statistic = {f_stat:.4f}, p-value = {p_value:.6f}")

if p_value < 0.05: # H0 = null hypothesis, that all algorithms perform the same
    print("We reject H0: At least one algorithm performs significantly differently at 95% confidence.")
else:
    print("We fail to reject H0: No statistically significant difference between algorithms at 95% confidence.")
```

## Model Submission for Project Step 2

```python
import cloudpickle
from sklearn.base import BaseEstimator, TransformerMixin

class NHANESCleaner(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
        df = X.copy()
        
        hiq_cols = ['HIQ032A', 'HIQ032B', 'HIQ032D', 'HIQ032F', 'HIQ032H', 'HIQ032I']
        for col in hiq_cols:
            if col in df.columns:
                df[col] = df[col].notna().astype(int)
        
        binary_remap = ['RIAGENDR', 'DMQMILIZ', 'DMDBORN4', 'ALQ111', 'ALQ151',
                        'HIQ011', 'HIQ210', 'INQ300']
        for col in binary_remap:
            if col in df.columns:
                df[col] = df[col].map({1: 1, 2: 0})
        
        cols_to_drop = ['SEQN', 'SDDSRVYR', 'RIDSTATR', 
                        'BPXOSY1', 'BPXODI1', 'LBXTC',
                        'BP', 'CHOL']
        df = df.drop(columns=[c for c in cols_to_drop if c in df.columns])
        
        return df
```

```python
nominal_cols = ['RIDRETH1', 'DMDMARTZ']
numeric_cols = ['RIAGENDR', 'RIDAGEYR', 'DMQMILIZ', 'DMDBORN4', 'DMDYRUSR',
                'DMDEDUC2', 'DMDHHSIZ', 'WTINT2YR', 'WTMEC2YR', 'ALQ111',
                'ALQ121', 'ALQ130', 'ALQ151', 'ALQ170', 'HIQ011', 'HIQ032A',
                'HIQ032B', 'HIQ032D', 'HIQ032F', 'HIQ032H', 'HIQ032I', 'HIQ210',
                'INDFMMPC', 'INQ300', 'IND310']

def make_feature_preprocessor():
    return ColumnTransformer([
        ('num', Pipeline([
            ('imputer', SimpleImputer(strategy='median')),
            ('scaler', StandardScaler())
        ]), numeric_cols),
        ('nom', Pipeline([
            ('imputer', SimpleImputer(strategy='most_frequent')),
            ('encoder', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
        ]), nominal_cols)
    ])
```

```python
nhanes_raw = pd.read_csv('nhanes.csv')
nhanes_raw['BP'] = nhanes_raw.apply(classify_bp, axis=1)

bp_train, _ = train_test_split(nhanes_raw, test_size=0.2, random_state=42)
bp_train = bp_train.dropna(subset=['BP'])

bp_X_sub = bp_train.drop(columns=['BP'])
bp_y_sub = bp_train['BP']

bp_submit_pipeline = Pipeline([
    ('cleaner', NHANESCleaner()),
    ('features', make_feature_preprocessor()),
    ('classifier', GradientBoostingClassifier(
        learning_rate=0.01, max_depth=3, min_samples_split=2,
        n_estimators=500, random_state=42
    ))
])

bp_submit_pipeline.fit(bp_X_sub, bp_y_sub)
```

```python
test_predictions = bp_submit_pipeline.predict(pd.read_csv('nhanes.csv').head())
print(test_predictions)
```
