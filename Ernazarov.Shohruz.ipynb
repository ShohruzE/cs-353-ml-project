### Task 3:

```python
# Create helper function to classify blood pressure for every row
def classify_bp(row):
    systolic_bp = row["BPXOSY1"]
    diastolic_bp = row["BPXODI1"]
    # Edge case where either systolic or diastolic are missing
    if pd.isna(systolic_bp) or pd.isna(diastolic_bp):
        return np.nan
    if systolic_bp < 120 and diastolic_bp < 80:
        return "normal"
    else:
        return "abnormal"
```

```python
# Pass in function to every row to classify and label the new BP column
nhanes['BP'] = nhanes.apply(classify_bp, axis=1)
nhanes['BP'].value_counts()
```

```python
nhanes['BP']
```

```python
def classify_cholesterol(value):
    if pd.isna(value):
        return np.nan
    if value < 200:
        return "normal"
    elif value <= 239:
        return "borderline"
    else:
        return "high"
```

```python
nhanes["CHOL"] = nhanes["LBXTC"].apply(classify_cholesterol)
nhanes["CHOL"].value_counts()
```

```python
nhanes = nhanes.drop(columns=['BPXOSY1', 'BPXODI1', 'LBXTC'])
```

### Task 5: 

```python
def pre_clean_data(df):
    
    # HIQ032 checkbox columns: NaN means "not checked" → 0, any value → 1
    hiq_cols = ['HIQ032A', 'HIQ032B', 'HIQ032D', 'HIQ032F', 'HIQ032H', 'HIQ032I']
    for col in hiq_cols:
        df[col] = df[col].notna().astype(int)
        
    # Remap binary 1/2 features to 1/0 (1=Yes stays 1, 2=No becomes 0)
    binary_remap = ['RIAGENDR', 'DMQMILIZ', 'DMDBORN4', 'ALQ111', 'ALQ151','HIQ011', 'HIQ210', 'INQ300']
    for col in binary_remap:
        df[col] = df[col].map({1: 1, 2: 0})

    # Drop non-predictive columns
    df = df.drop(columns=['SEQN', 'SDDSRVYR', 'RIDSTATR'])
    
    return df
```

```python
bp_clean = pre_clean_data(bp_training_ernazarov)
chol_clean = pre_clean_data(chol_training_ernazarov)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer

nominal_cols = ['RIDRETH1', 'DMDMARTZ']

nominal_pipeline = Pipeline(steps=[
    ('impute', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

preprocessor_b = ColumnTransformer(
    transformers=[
        ('nominal', nominal_pipeline, nominal_cols)
])

preprocessor_c = ColumnTransformer(
    transformers=[
        ('nominal', nominal_pipeline, nominal_cols)
])

plineb = Pipeline(steps=[('preprocessor', preprocessor_b)])
plinec = Pipeline(steps=[('preprocessor', preprocessor_c)])
```

```python
bp_X = bp_clean.drop(columns=['BP', 'CHOL'])
bp_y = bp_clean['BP']
bp_transformed = plineb.fit_transform(bp_X)

bp_feature_names = plineb.named_steps['preprocessor'].get_feature_names_out()
bp_transformed_df = pd.DataFrame(bp_transformed, columns=bp_feature_names)
bp_transformed_df.tail()
```

```python
chol_X = chol_clean.drop(columns=['BP', 'CHOL'])
chol_y = chol_clean['CHOL']

chol_transformed = plinec.fit_transform(chol_X)

chol_feature_names = plinec.named_steps['preprocessor'].get_feature_names_out()
chol_transformed_df = pd.DataFrame(chol_transformed, columns=chol_feature_names)
chol_transformed_df.tail()
```
